```python
sns.displot(
    data=property_plots_df,
    x="price",
    kind="hist",
    aspect=1.4,
    log_scale=10,
    bins=20
)
```

```python
property_plots_df['price'].round(5)
```

```python
column_grade = property_plots_df["kcgs_grade"]
column_condition = property_plots_df["kcgs_condition"]

correlation_cond_gr = column_grade.corr(column_condition)
print(correlation_cond_gr)
```

```python
property_plots_df.corr()
```

```python
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(property_plots_df.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':8}, pad=12)

plt.show()
```
