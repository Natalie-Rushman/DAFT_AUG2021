```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
breast_cancer = pd.read_csv('breast-cancer.csv', header=None)
breast_cancer.head()
```

```python
# renaming the columns
total_cols = ['id', 'outcome', 'time', 'cell_1_radius', 'cell_1_texture', 'cell_1_perimiter',
              'cell_1_area', 'cell_1_smoothness', 'cell_1_compactness', 'cell_1_concavity',
              'cell_1_concave_points', 'cell_1_symmetry', 'cell_1_fractal_dimension',
              'cell_2_radius', 'cell_2_texture', 'cell_2_perimiter', 'cell_2_area',
              'cell_2_smoothness', 'cell_2_compactness', 'cell_2_concavity',
              'cell_2_concave_points', 'cell_2_symmetry', 'cell_2_fractal_dimension', 'cell_3_radius', 
              'cell_3_texture', 'cell_3_perimiter', 'cell_3_area', 'cell_3_smoothness',
              'cell_3_compactness', 'cell_3_concavity', 'cell_3_concave_points', 'cell_3_symmetry',
              'cell_3_fractal_dimension', 'tumor_size', 'lymph_status']
breast_cancer.columns = total_cols
```

```python
breast_cancer.head()
```

```python
# getting the numerical data
numerics = breast_cancer._get_numeric_data()
numerics.head()
```

```python
numerics = numerics.drop(['id','time'], axis=1)
```

```python
# splitting target variable from independent variables
X = numerics
y = breast_cancer["outcome"]
```

```python
from sklearn.preprocessing import StandardScaler
transformer = StandardScaler().fit(X)
scaled = transformer.transform(X)
scaled = pd.DataFrame(scaled)
scaled.columns = numerics.columns 
scaled.head()
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
pca = PCA()
pca.fit(scaled)
```

```python
pca.components_ ### coefficients
```

```python
### variations = pca.explained_variance_ ### How many components are needed?
variations = pca.explained_variance_ratio_*100
variations = np.round(variations, decimals=1)
variations
```

```python
variances = pca.explained_variance_ratio_.tolist()
```

```python
### displaying the first two (co-)variances
variances[:2]
```

```python
### table feat. variances and cumulated variances
total_variance = [sum(variances[:i]) for i in range(1, len(variances)+1)]
scree = pd.DataFrame({'nbr_pca': range(1,len(variances)+1),'variance': variances,'cum_variance': total_variance})
scree.head()
```

```python
plot_labels = ['PC'+str(i) for i in range(1,len(variations)+1)]
plot_labels
```

```python
from matplotlib.pyplot import figure
figure(figsize=(20,5))
plt.bar(x = plot_labels, height=variations)
plt.xlabel('Principal Components')
plt.ylabel('Variance Explained')
plt.title('PCA Scree Plot')
```

```python
### deciding on the optimal amount of principal components
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_style("darkgrid", {"axes.facecolor": ".9"})

fig, ax = plt.subplots(1,2,figsize=(16,8))

sns.lineplot(x="nbr_pca", y="variance", data=scree, ax = ax[0], color = 'black')
ax[0].set_xticks(range(1,32))
sns.barplot(x="nbr_pca", y="cum_variance", data=scree, ax = ax[1], color = 'lightblue')
```
