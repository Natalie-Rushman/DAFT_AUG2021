## Write a function to replace column "State" to different zones:
- California = West Region
- Oregon = North West
- Washington = East
- Arizona, Nevada = Central

```python
df.State.value_counts()
```

```python
### mapping >>> dictionary

def state_group(x):
    if not x == x:
        return x
    else:
        dct = {'Washington' : 'East',
              'Arizona' : 'Central',
              'California' : 'West',
               'Oregon' : 'North West',
               'Nevada' : 'Central'
              }
        return dct[x]

df['Region'] = df['State'].apply(state_group)
```

```python
df['region_mapped'] = list(map(state_group, df['State']))
```

```python
df['region_mapped'].unique()
```
